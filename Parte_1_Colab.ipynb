## Estandarizar fechas

````python
def cambio(x):
  y=""
  for letra in x:
    if letra=="/":
      letra="-"
    y+=letra
  c=""
  a=y.split()
  b=a[0].split("-")
  if len(b[0])<2 and len(b[1])<2 and len(b[2])<3:
    c+="0"+b[0]+"-"+"0"+b[1]+"-"+"20"+b[2]+" "+"00:00"
  elif len(b[0])<2 and len(b[1])<2:
    c+="0"+b[0]+"-"+"0"+b[1]+"-"+b[2]+" "+"00:00"
  elif len(b[0])<2 and len(b[2])<3:
    c+="0"+b[0]+"-"+b[1]+"-"+"20"+b[2]+" "+"00:00"
  elif len(b[1])<2 and len(b[2])<3:
    c+=b[0]+"-"+"0"+b[1]+"-"+"20"+b[2]+" "+"00:00"
  elif len(b[0])<2:
    c+="0"+b[0]+"-"+b[1]+"-"+b[2]+" "+"00:00"
  elif len(b[1])<2:
    c+=b[0]+"-"+"0"+"-"+b[1]+" "+"00:00"
  elif len(b[2])<3:
    c+=b[0]+"-"+b[1]+"-"+"20"+b[2]+" "+"00:00"
  else:
    c+=b[0]+"-"+b[1]+"-"+b[2]+" "+"00:00"
  return c

data=pd.read_csv('data-iic1005-2019.csv',converters={"fecha_siniestro": cambio
                                                  })
data
````

## Gráfico de línea sobre cantidad de robos según día de la semana

````python

data["fecha_siniestro"] = pd.to_datetime(data.fecha_siniestro, format='%d-%m-%Y %H:%M')
````

````python
pd.value_counts(data.fecha_siniestro.dt.weekday_name)
````

````python
Datos={"Dias de la semana":["GDomingo","FSabado","EViernes","DJueves","CMiercoles","BMartes","ALunes"],"Frecuencia de robos": ["155","186","241","250","254","247","225"]}
df=pd.DataFrame(Datos)
df=df.sort_values(by="Frecuencia de robos",ascending=True)
df
````

````python
print("Gráfico cantidad de robos por día de semana")

source = pd.DataFrame({
  'x': Datos["Dias de la semana"],
  'f(x)': Datos["Frecuencia de robos"]},
)
alt.Chart(source).mark_line().encode(
    x='x',
    y='f(x):Q', 
)
````

## Gráfico de barra mes versus cantidad de robos

````python
pd.value_counts(data.fecha_siniestro.dt.month)
dato_meses={"Meses":["aEnero","bFebrero","cMarzo","dAbril","eMayo","fJunio","gJulio","hAgosto","iSeptiembre","jOctubre","kNoviembre","lDiciembre"],"Cantidad de Robos": ["263","240","324","293","411","23","1","0","0","1","0","2"]}
pm=pd.DataFrame(dato_meses)
pm
````

````python
print("Gráfico Horizontal cantidad de robos por mes")
alt.Chart(pm).mark_bar().encode(
    alt.Y("Meses:O"),
    alt.X('Cantidad de Robos:Q'),
    color='Meses',
    
).interactive(bind_x=False)
````
